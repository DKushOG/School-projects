#### 2.1.1 Toyota Data Collection

<p><b>Singapore Toyota car pricing lists:</b></p>
<p>The data taken will be extracted from price lists in PDF format from reputable websites such as SG Car Mart. For consitency and due to the availability of readable data from this source, the price list data gathered will be from 2 seperate months in the aforementioned time period to be able to draw a reasonable average for each half of the years. (Oct 2023, Nov 2023, Jan 2024 and June 2024).</p>
<p>The code below shows the method by which the data will be extracted for Toyota price lists.</p>

```python
import PyPDF2

def extract_text_from_pdf(pdf_path, txt_path):
    """
    Extracts text from a PDF file and saves it to a text file.
    
    Args:
    - pdf_path: Path to the PDF file.
    - txt_path: Path to save the extracted text as a text file.
    """
    try:
        # Open the PDF file
        with open(pdf_path, 'rb') as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            
            # Extract text
            all_text = ''
            for page in reader.pages:
                all_text += page.extract_text() + '\n'
            
            # Save the extracted text to a file
            with open(txt_path, 'w', encoding='utf-8') as txt_file:
                txt_file.write(all_text)
                
        print(f"Text extracted and saved to {txt_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the PDF files and output text files
pdf_files = [
    (r'DataSources\Toyota2024JunePrice.pdf', r'Output\Toyota2024JunePrice.txt'),
    (r'DataSources\Toyota2024JanPrice.pdf', r'Output\Toyota2024JanPrice.txt'),
    (r'DataSources\Toyota2023NovPrice.pdf', r'Output\Toyota2023NovPrice.txt'),
    (r'DataSources\Toyota2023OctPrice.pdf', r'Output\Toyota2023OctPrice.txt')
]

# Extract text from each PDF
for pdf_path, txt_path in pdf_files:
    extract_text_from_pdf(pdf_path, txt_path)
```

<p>Due to the complex formatting of the data in the Toyota pricing lists, the pdf text had to be extracted into a text file to be able to be parsed properly.</p>
<p>Below shows the processing of the output .txt files into a readable CSV file.</p>

```python
import csv
import re

def extract_pricing_data(file_path, date_range):
    """
    Extracts pricing data for all car categories from a Toyota pricing text file.
    Args:
    - file_path: Path to the text file.
    - date_range: Date range from the text file (used for tagging data).
    
    Returns:
    - A list of dictionaries containing the extracted pricing data.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    
    # Match vehicle details and their pricing
    matches = re.findall(
        r'([\w\s&]+)\s+([\w\d\-\/\(\)]+)\s+([\w\d\-\/\(\)]+)\s+([\d,]+)\s+\$\s+([\d,]+)\s+\$',
        text
    )
    
    data = []
    for match in matches:
        category = match[0].strip()
        model = match[1].strip()
        variant = match[2].strip()
        without_coe_price = match[3].replace(",", "")
        with_coe_price = match[4].replace(",", "")
        
        data.append({
            "Category": category,
            "Model": model,
            "Variant": variant,
            "Without COE Price": without_coe_price,
            "With COE Price": with_coe_price,
            "Year": date_range
        })
    return data

# Paths to the text files and their respective date ranges
text_files = {
    r"Output\Toyota2023NovPrice.txt": "2023",
    r"Output\Toyota2023OctPrice.txt": "2023",
    r"Output\Toyota2024JanPrice.txt": "2024",
    r"Output\Toyota2024JunePrice.txt": "2024"
}

# Extract data from each file
all_data = []
for file_path, date_range in text_files.items():
    all_data.extend(extract_pricing_data(file_path, date_range))

# Write consolidated data to a CSV file
output_csv_path = r"Output\Toyota_pricing_consolidated.csv"
with open(output_csv_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ["Category", "Model", "Variant", "Without COE Price", "With COE Price", "Year"]
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(all_data)

print(f"Pricing data has been consolidated into {output_csv_path}.")
```

<p>This output consolidates the prices of the various car models by price and model into one CSV file for further analysis. As Toyota's SG pricelists follow a similar format, this could be used for more detailed analysis int the future.</p>

<p><b>Toyota UK car pricing lists:</b></p>
<p>Toyota UK releases their price lists in PDF formats on their website. However the only releases I could find were from 2023 May and 2024 April respectively. These shara a similar unconventional format, therefore they will also have to be extracted into text files initally.</p>
<p>Below shows the code used to extract the data into text files.</p>

```python
from PyPDF2 import PdfReader

def convert_pdf_to_text(pdf_path, output_txt_path):
    """
    Converts a PDF file into a text file.
    
    Args:
    - pdf_path: Path to the PDF file.
    - output_txt_path: Path to the output text file.
    """
    try:
        # Open the PDF file
        reader = PdfReader(pdf_path)
        with open(output_txt_path, 'w', encoding='utf-8') as txt_file:
            for page_num, page in enumerate(reader.pages, start=1):
                text = page.extract_text()
                txt_file.write(f"--- Page {page_num} ---\n")
                txt_file.write(text + "\n\n")
        print(f"PDF converted to text and saved at {output_txt_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# List of PDFs and their corresponding output text file paths
pdf_to_text_files = {
    r"DataSources\ToyotaUK2023MayPrice.pdf": r"Output\ToyotaUK2023MayPrice.txt",
    r"DataSources\ToyotaUK2024AprilPrice.pdf": r"Output\ToyotaUK2024AprilPrice.txt",
}

# Convert each PDF to text
for pdf_path, output_txt_path in pdf_to_text_files.items():
    convert_pdf_to_text(pdf_path, output_txt_path)
```

<p>Below shows the processing of the output .txt files into a readable CSV file:</p>

```python
import re
import os
import pandas as pd

# Define a function to extract data with regex based on the year
def extract_car_data(file_path, file_year):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    data = []

    # Use a different regex pattern for 2024 data
    if file_year == 2023:
        regex = r"([A-Za-z0-9\-\'\s]+)\nFrom\s+£([\d,]+\.\d{2})"
    elif file_year == 2024:
        regex = r"([A-Za-z0-9\-\'\s]+)[^\d]*£[\d,]+\.\d{2}\s+£([\d,]+\.\d{2})"
    else:
        print(f"No regex defined for year {file_year}. Skipping file.")
        return data

    matches = re.findall(regex, text)
    for model, price in matches:
        data.append({
            "Year": file_year,
            "Model": model.strip(),
            "Total Price": float(price.replace(',', ''))
        })

    return data

# Extract year from the file name
def extract_year_from_filename(filename):
    match = re.search(r"(\d{4})", filename)
    return int(match.group(1)) if match else None

# File paths
file_paths = [
    r"Output\ToyotaUK2023MayPrice.txt",
    r"Output\ToyotaUK2024AprilPrice.txt"
]

# Aggregate data from all files
all_data = []
for file_path in file_paths:
    file_name = os.path.basename(file_path)
    file_year = extract_year_from_filename(file_name)
    if file_year:
        extracted_data = extract_car_data(file_path, file_year)
        if extracted_data:
            all_data.extend(extracted_data)
    else:
        print(f"Year not found in file name: {file_name}")

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Save to CSV
output_path = r"Output\Toyota_Car_Prices_UK_Fixed.csv"
df.to_csv(output_path, index=False)

print(f"Data extraction completed successfully. CSV saved to: {output_path}")
```

#### 2.2.1 Singapore Data
<p>As mentioned, data has been provided from Singapore's LTA in the form of PDF files that detail the car make and total new registrations for the year.</p>
<p>The following code is used to parse the data and format it into a CSV file:</p>

```python
import PyPDF2
import re
import pandas as pd

# Function to extract Toyota data from a PDF
def extract_toyota_data(file_path, year):
    data = []
    with open(file_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        for page in reader.pages:
            text = page.extract_text()
            # Find all Toyota registration numbers using regex
            matches = re.findall(r'TOYOTA\s+.*?\s+(\d+(?:,\d+)*)', text)
            for match in matches:
                if match:
                    data.append({"Year": year, "Brand": "Toyota", "Total Registrations": int(match.replace(",", ""))})
    return data

# File paths and years
file_2023 = r"DataSources\2023SgCarOwnership.pdf"
file_2024 = r"DataSources\2024SgCarOwnership.pdf"
year_2023 = 2023
year_2024 = 2024

# Extract data for each year
data_2023 = extract_toyota_data(file_2023, year_2023)
data_2024 = extract_toyota_data(file_2024, year_2024)

# Combine data
data_combined = data_2023 + data_2024

# Convert to DataFrame
df = pd.DataFrame(data_combined)

# Save to CSV
output_file = r"Output\Toyota_Car_Registrations_SG.csv"
df.to_csv(output_file, index=False)

print(f"Toyota data saved to {output_file}")
```

#### 2.2.2 UK Data

<p>For car registrations by make in the UK, I have found a table detailing the new registrations of vehicles per make similar to Singapore LTA's data tables.</p>
<p>The following code webscrapes the publicly available data into a CSV file:</p>

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the target page
url = "https://www.smmt.co.uk/vehicle-data/car-registrations/"

# Send a GET request
response = requests.get(url)
response.raise_for_status()  # Check for request errors

# Parse the page content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Define a function to extract data for specific brands
def extract_brand_data(soup, brands):
    data = []

    # Locate the specific table
    table = soup.find("table", width="488")  # Specific table based on attributes
    if not table:
        print("Table not found on the page.")
        return data

    # Extract table rows
    rows = table.find_all("tr")

    for row in rows[2:]:  # Skip header rows
        cells = row.find_all("td")
        if len(cells) >= 5:  # Ensure there are enough cells in the row
            brand = cells[0].get_text(strip=True)
            data_2024 = cells[1].get_text(strip=True).replace(",", "")
            data_2023 = cells[3].get_text(strip=True).replace(",", "")

            if brand in brands:
                data.append({"Year": "2024", "Brand": brand, "Number Registered": int(data_2024)})
                data.append({"Year": "2023", "Brand": brand, "Number Registered": int(data_2023)})

    return data

# Specify the brands to extract
brands_of_interest = ["Toyota"]

# Extract the data
data = extract_brand_data(soup, brands_of_interest)

# Convert to a DataFrame
df = pd.DataFrame(data)

# Save to a CSV file
output_file = r"Comparison\UK_car_registrations.csv"
df.to_csv(output_file, index=False)
```

### 3.1 Vehicle Pricing
<p>Due to the variations in data provided and the unavoidable variance of output formats given by previous code, i will need to process the output CSV files further to appropriate the data for graphing and comparisons.</p>

```python
import pandas as pd

# Load UK and Singapore data
uk_file_path = r"Output\Toyota_Car_Prices_UK_Fixed.csv"
sg_file_path = r"Output\Toyota_pricing_consolidated.csv"

uk_data = pd.read_csv(uk_file_path)
sg_data = pd.read_csv(sg_file_path)

# Filter UK data: Remove entries with prices below 10,000 GBP
uk_data_filtered = uk_data[uk_data['Total Price'] >= 10000]

# Include both 2023 and 2024 entries
uk_data_2023_2024 = uk_data_filtered[uk_data_filtered['Year'].isin([2023, 2024])]

# Define a function to categorize price into tiers
def categorize_price(price):
    if price < 25000:
        return 'Low'
    elif 25001 <= price <= 40000:
        return 'Mid'
    else:
        return 'High'

# Apply the categorization to UK data
uk_data_2023_2024['Price_Tier'] = uk_data_2023_2024['Total Price'].apply(categorize_price)

# Add Location column for UK data
uk_data_2023_2024['Location'] = 'UK'

# Define a function to categorize Singapore car prices
def categorize_sg_price(price):
    if price < 200000:
        return 'Low'
    elif 200001 <= price <= 300000:
        return 'Mid'
    else:
        return 'High'

# Apply categorization to Singapore data
sg_data['Price_Tier'] = sg_data['With COE Price'].apply(categorize_sg_price)

# Add Location column for Singapore data
sg_data['Location'] = 'Singapore'

# Concatenate both datasets
consolidated_data = pd.concat([
    uk_data_2023_2024.assign(Average_Price_SGD=pd.NA),
    sg_data.assign(Average_Price_GBP=pd.NA)
], ignore_index=True)

# Save the corrected consolidated data
output_path = r"Comparison\Toyota_Consolidated_Data_Updated.csv"
consolidated_data.to_csv(output_path, index=False)

print(f"Consolidated data saved to {output_path}")
```

<p>This code has grouped the vehicle data into: The grouping of data aims to provide a clear and actionable categorization of Toyota car models based on their average price ranges in two different markets: the UK and Singapore. By calculating and classifying models into tiers (Low, Mid, High), the consolidation simplifies comparisons and allows for highlights pricing trends in each region. For the UK, tiers are defined in pounds, reflecting affordability and market segmentation, while in Singapore, the focus is on "With COE Price," reflecting the unique cost structure of vehicle ownership. </p>
<p>However, the data needs to be further refined and cleaned for other planned analysis:</p>

```python
import pandas as pd

# Load the updated consolidated data
consolidated_file_path = r"Comparison\Toyota_Consolidated_Data_Updated.csv"
consolidated_data = pd.read_csv(consolidated_file_path)

# Calculate averages for each tier by year and location
average_prices = consolidated_data.groupby(['Year', 'Location', 'Price_Tier']).agg(
    Avg_Price_GBP=('Total Price', 'mean'),
    Avg_With_COE_Price=('With COE Price', 'mean')
).reset_index()

# Rename columns to match the desired format
average_prices.rename(columns={
    'Location': 'Country',
    'Price_Tier': 'Tier',
    'Avg_Price_GBP': 'Avg Price GBP',
    'Avg_With_COE_Price': 'Avg With COE Price'
}, inplace=True)

# Save the result to a new CSV file
output_file_path = r"Comparison\Toyota_Average_Prices_By_Tier.csv"
average_prices.to_csv(output_file_path, index=False)

print(f"Averages by tier saved to {output_file_path}")
```

<p>To facilitate meaningful comparisons and visualizations, the data was refined by averaging the respective price tiers (Low, Mid, High) for each year and country. This process involved grouping the data by year, country, and tier, then calculating the mean values for the average prices in GBP and the average prices inclusive of COE. By consolidating these averages, we obtained a simplified dataset that highlights the pricing trends and tier distributions across different regions and time periods, enabling clearer insights for graphing and analysis.</p>

### 4.2 Analyzing Pricing Tiers Across Regions (Bar Charts)
<p>Bar charts are a good way to distinguish trends in the data gathered to compare average prices across pricing tiers for the UK and Singapore.</p>
<p>Below shows a clustered bar graph representing the pricing trends.</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the input data
file_path = r"Comparison\Toyota_Average_Prices_By_Tier.csv"
data = pd.read_csv(file_path)

# Define the desired order of tiers
tier_order = ['Low', 'Mid', 'High']

# Prepare data for UK and Singapore with year included
uk_data = data[data['Country'] == 'UK']
sg_data = data[data['Country'] == 'Singapore']

# Reorganize data to ensure tiers are ordered and years are considered
uk_data = uk_data.pivot(index='Tier', columns='Year', values='Avg Price GBP').reindex(tier_order)
sg_data = sg_data.pivot(index='Tier', columns='Year', values='Avg With COE Price').reindex(tier_order)

# Create the bar charts
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Colors for years
colors = plt.cm.tab10(np.linspace(0, 1, len(data['Year'].unique())))

# Bar width and positions
bar_width = 0.2
x_positions = np.arange(len(tier_order))

# UK bar chart
uk_years = uk_data.columns
for i, year in enumerate(uk_years):
    axes[0].bar(x_positions + i * bar_width, uk_data[year], bar_width, label=f"{year}", color=colors[i])

axes[0].set_xlabel("Price Tier")
axes[0].set_ylabel("Average Price (GBP)")
axes[0].set_title("Average Prices by Tier in the UK")
axes[0].set_xticks(x_positions + bar_width * (len(uk_years) - 1) / 2)
axes[0].set_xticklabels(tier_order)
axes[0].legend(title="Year")

# Singapore bar chart
sg_years = sg_data.columns
for i, year in enumerate(sg_years):
    axes[1].bar(x_positions + i * bar_width, sg_data[year], bar_width, label=f"{year}", color=colors[i])

axes[1].set_xlabel("Price Tier")
axes[1].set_ylabel("Average Price (With COE, SGD)")
axes[1].set_title("Average Prices by Tier in Singapore")
axes[1].set_xticks(x_positions + bar_width * (len(sg_years) - 1) / 2)
axes[1].set_xticklabels(tier_order)
axes[1].legend(title="Year")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

    <ul>

        <li>This gradual increase reflects the differentiation in features and luxury levels among the tiers, consistent with consumer expectations for value-added attributes in higher-priced vehicles.</li>

     <li><b>Singapore Pricing Trends:</b></li>
    <ul>

### 4.3 Visualizing Combined Trends (Dual-Axis Bar-Line Chart)

<p>Given both data trends, combining them into a Dual-Axis chart will help to provide insights into the correlation between the data gathered.</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the input data
file_path = r"Comparison\Toyota_Average_Prices_By_Tier.csv"
sg_registrations_path = r"Comparison\Aggregated_Toyota_Registrations_SG.csv"
uk_registrations_path = r"Comparison\UK_car_registrations.csv"

data = pd.read_csv(file_path)
sg_registrations = pd.read_csv(sg_registrations_path)
uk_registrations = pd.read_csv(uk_registrations_path)

# Process registration data
uk_registrations = uk_registrations[uk_registrations['Brand'] == 'Toyota'][['Year', 'Number Registered']]
uk_registrations.rename(columns={'Number Registered': 'Total Registrations'}, inplace=True)
sg_registrations = sg_registrations[['Year', 'Total Registrations']]

registrations = pd.concat([
    uk_registrations.assign(Country='UK'),
    sg_registrations.assign(Country='Singapore')
], ignore_index=True)

# Define tier order and prepare data for bar chart
tier_order = ['Low', 'Mid', 'High']
uk_price_data = data[data['Country'] == 'UK']
sg_price_data = data[data['Country'] == 'Singapore']

uk_price_pivot = uk_price_data.pivot(index='Tier', columns='Year', values='Avg Price GBP').reindex(tier_order)
sg_price_pivot = sg_price_data.pivot(index='Tier', columns='Year', values='Avg With COE Price').reindex(tier_order)

# Create the combined chart
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

colors = plt.cm.tab10(np.linspace(0, 1, len(data['Year'].unique())))
line_colors = ['red', 'orange']
bar_width = 0.2
x_positions = np.arange(len(tier_order))

# UK Chart
uk_years = uk_price_pivot.columns
for i, year in enumerate(uk_years):
    axes[0].bar(x_positions + i * bar_width, uk_price_pivot[year], bar_width, label=f"{year}", color=colors[i])

uk_registration_trend = registrations[registrations['Country'] == 'UK']
axes_uk_line = axes[0].twinx()
uk_line_values = [uk_registration_trend.loc[uk_registration_trend['Year'] == year, 'Total Registrations'].values[0] for year in uk_years]
axes_uk_line.plot(
    [x_positions[0] - bar_width, x_positions[-1] + bar_width * len(uk_years)], 
    uk_line_values, 
    color=line_colors[0], marker='o', linestyle='-', label="Total Registrations"
)

# Add year labels to the points
for i, (x, y) in enumerate(zip(x_positions + bar_width * len(uk_years) / 2, uk_line_values)):
    # Apply an offset for the 2024 label
    if uk_years[i] == 2024:
        x += 1.6  # Shift the label slightly to the right
    axes_uk_line.text(x - 0.4, y + 60, str(uk_years[i]), ha='center', fontsize=9)



# Adjust y-axis scale for more space
axes_uk_line.set_ylim(min(uk_line_values) - 2000, max(uk_line_values) + 2000)

axes[0].set_xlabel("Price Tier")
axes[0].set_ylabel("Average Price (GBP)")
axes[0].set_title("Average Prices and Registrations in the UK")
axes[0].set_xticks(x_positions + bar_width * (len(uk_years) - 1) / 2)
axes[0].set_xticklabels(tier_order)
axes_uk_line.set_ylabel("Total Registrations")
axes[0].legend(title="Year", loc="upper left")
axes_uk_line.legend(loc="upper right")

# Singapore Chart
sg_years = sg_price_pivot.columns
for i, year in enumerate(sg_years):
    axes[1].bar(x_positions + i * bar_width, sg_price_pivot[year], bar_width, label=f"{year}", color=colors[i])

sg_registration_trend = registrations[registrations['Country'] == 'Singapore']
axes_sg_line = axes[1].twinx()
sg_line_values = [sg_registration_trend.loc[sg_registration_trend['Year'] == year, 'Total Registrations'].values[0] for year in sg_years]
axes_sg_line.plot(
    [x_positions[0] - bar_width, x_positions[-1] + bar_width * len(sg_years)], 
    sg_line_values, 
    color=line_colors[1], marker='o', linestyle='-', label="Total Registrations"
)

# Add year labels to the points
# Add year labels to the points
# Add year labels to the points
for i, (x, y) in enumerate(zip(x_positions + bar_width * len(sg_years) / 2, sg_line_values)):
    # Apply an offset for the 2024 label
    if sg_years[i] == 2024:
        x += 1.6  # Shift the label slightly to the right
    axes_sg_line.text(x - 0.4, y + 60, str(sg_years[i]), ha='center', fontsize=9)


# Adjust y-axis scale for more space
axes_sg_line.set_ylim(min(sg_line_values) - 2000, max(sg_line_values) + 2000)

axes[1].set_xlabel("Price Tier")
axes[1].set_ylabel("Average Price (With COE, SGD)")
axes[1].set_title("Average Prices and Registrations in Singapore")
axes[1].set_xticks(x_positions + bar_width * (len(sg_years) - 1) / 2)
axes[1].set_xticklabels(tier_order)
axes_sg_line.set_ylabel("Total Registrations")
axes[1].legend(title="Year", loc="upper left")
axes_sg_line.legend(loc="upper right")

plt.tight_layout()
plt.show()
```
